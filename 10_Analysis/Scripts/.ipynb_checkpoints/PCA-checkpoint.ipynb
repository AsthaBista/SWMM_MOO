### **Import Packages & Load Data**

```python
# Import Packages & Load Data
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv(r'C:\Users\ABI\OneDrive - NIVA\Documents\GitHub\SWMM_MOO\10_Analysis\1002_PCA\1002_All_objectives.csv', sep=';', index_col='sim')
print("Data Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

### **Data Cleaning & Inspection**

```python
pollutant_columns = ['CO2', 'TSS', 'TP', 'TN', 'Cu', 'Pb', 'Zn']
print("Data for sim42 (the outlier):")
print(df.loc['sim42', pollutant_columns])

df_clean = df.copy()

for col in pollutant_columns:
    upper_cap = df_clean[col].quantile(0.99)
    lower_cap = df_clean[col].quantile(0.01)
    df_clean[col] = df_clean[col].clip(lower=lower_cap, upper=upper_cap)

print("\nCapped value for sim42 TSS:", df_clean.loc['sim42', 'TSS'])
print("Capped value for sim42 CO2:", df_clean.loc['sim42', 'CO2'])
```

### **Perform PCA on Selected Objectives**

A correlation analysis was initially carried out in excel. One objective was selected if two or more are highly correlated. 
Then seven objectives were selected that did not have high correlation with one another

```python
selected_objectives = ['PR10', 'TR', 'BGF', 'Inv', 'UNA', 'CO2', 'TSS']
print(f"\nPerforming PCA on selected objectives: {selected_objectives}")

X_selected = df_clean[selected_objectives].copy()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_selected)

pca = PCA()
pca.fit(X_scaled)
X_pca = pca.transform(X_scaled)

loadings_df = pd.DataFrame(pca.components_.T,
                           columns=[f'PC{i+1}' for i in range(pca.n_components_)],
                           index=selected_objectives)

print("\nExplained Variance Ratio for each PC:")
print(pd.Series(pca.explained_variance_ratio_, index=loadings_df.columns))
print("\nCumulative Explained Variance:")
print(np.cumsum(pca.explained_variance_ratio_))

print("\n" + "="*70)
print("PCA LOADINGS TABLE")
print("="*70)
loadings_df
```

### **Analyze Dominant Objectives**

```python
print("\n" + "="*60)
print("DOMINANT OBJECTIVES PER PRINCIPAL COMPONENT")
print("="*60)

n_components_to_show = min(4, len(selected_objectives))

for i in range(n_components_to_show):
    pc_name = f'PC{i+1}'
    pc_loadings = loadings_df[pc_name].abs().sort_values(ascending=False)
    dominant_mask = pc_loadings > 0.4
    dominant_objs = pc_loadings[dominant_mask]
    
    print(f"\n{pc_name} (Explains {pca.explained_variance_ratio_[i]:.1%} of variance):")
    if not dominant_objs.empty:
        for obj, loading in dominant_objs.items():
            original_loading = loadings_df.loc[obj, pc_name]
            print(f"  - {obj} (Loading: {original_loading:.3f})")
    else:
        top_obj = pc_loadings.index[0]
        top_loading = loadings_df.loc[top_obj, pc_name]
        print(f"  - {top_obj} (Loading: {top_loading:.3f})")
```

##### Selection of dominant objectives

The variance in the data is quite distributed amongst the PCs, with PC1 explaining only 42% of variance. Therefore, other PCs are also important. Two PCs explain only 58% variance, and three PCs explain 72%. Here four Pcs explaining total of 86% of variance is selected for further analysis.

From the above list, it can be observed that in PC1, there are BGF, Inv, and CO2 that have high loadings. BGF and Inv are correlated with each other, therefore taking only one from PC1, say, BGF.
Taking TSS from PC2.
Taking PR10, and TR, they have high variance in PC3 and PC4. 

```python
selected_for_plot = ['BGF', 'TSS', 'PR10', 'TR']
loadings_subset = loadings_df.loc[selected_for_plot]

fig = plt.figure(figsize=(18, 5))

# Plot 1: PC1 vs PC2
ax1 = fig.add_subplot(131)
for idx, row in loadings_subset.iterrows():
    ax1.arrow(0, 0, row['PC1'], row['PC2'], head_width=0.05, head_length=0.05, fc='k', ec='k')
    ax1.text(row['PC1']*1.15, row['PC2']*1.15, idx, color='darkred', ha='center', va='center', fontweight='bold')
ax1.axhline(y=0, color='k', linestyle='--', alpha=0.5)
ax1.axvline(x=0, color='k', linestyle='--', alpha=0.5)
ax1.set_xlim(-1, 1)
ax1.set_ylim(-1, 1)
ax1.set_xlabel('Principal Component 1')
ax1.set_ylabel('Principal Component 2')
ax1.set_title('PCA Loadings Plot: PC1 vs PC2')
ax1.grid(True)

# Plot 2: PC3 vs PC4
ax2 = fig.add_subplot(132)
for idx, row in loadings_subset.iterrows():
    ax2.arrow(0, 0, row['PC3'], row['PC4'], head_width=0.05, head_length=0.05, fc='k', ec='k')
    ax2.text(row['PC3']*1.15, row['PC4']*1.15, idx, color='darkblue', ha='center', va='center', fontweight='bold')
ax2.axhline(y=0, color='k', linestyle='--', alpha=0.5)
ax2.axvline(x=0, color='k', linestyle='--', alpha=0.5)
ax2.set_xlim(-1, 1)
ax2.set_ylim(-1, 1)
ax2.set_xlabel('Principal Component 3')
ax2.set_ylabel('Principal Component 4')
ax2.set_title('PCA Loadings Plot: PC3 vs PC4')
ax2.grid(True)

# Plot 3: 3D Plot
ax3 = fig.add_subplot(133, projection='3d')
for idx, row in loadings_subset.iterrows():
    ax3.quiver(0, 0, 0, row['PC1'], row['PC2'], row['PC3'], color='k', arrow_length_ratio=0.05)
    ax3.text(row['PC1']*1.15, row['PC2']*1.15, row['PC3']*1.15, idx, color='green', fontweight='bold')
ax3.set_xlim(-1, 1)
ax3.set_ylim(-1, 1)
ax3.set_zlim(-1, 1)
ax3.set_xlabel('PC1')
ax3.set_ylabel('PC2')
ax3.set_zlabel('PC3')
ax3.set_title('3D PCA Loadings Plot')

plt.tight_layout()
plt.savefig(r'C:\Users\ABI\OneDrive - NIVA\Documents\GitHub\SWMM_MOO\10_Analysis\\1002_PCA\PCA_Loadings_Plot.png', 
            dpi=300, 
            bbox_inches='tight')
plt.show()
```

The variance in these objectives are quite distributed. Therefore, these objectives are selected for further analysis.
